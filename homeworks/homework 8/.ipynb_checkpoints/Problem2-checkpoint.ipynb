```python
import numpy as np
import pandas as pd
from astropy.table import Table
import matplotlib.pyplot as plt
```

```python
#1 Read in File
tbl = Table.read('/Users/brandummb/Downloads/global_CCl4_MM.dat', format='ascii')
```

```python
#2 Convert to Pandas
Date = []
GLm = []
GLsd = []
for row in tbl:
    date_float = (float(row['CCl4ottomon']) - 1) / 12 + float(row['CCl4ottoyr'])
    Date.append(date_float)
    GLm.append(row['CCl4ottoGLm'])
    GLsd.append(row['CCl4ottoGLsd'])
    
df = pd.DataFrame({'Date': Date, 'Global Mean Concentration': GLm, 'Global Mean Concentration Standard Deviation': GLsd})

df.astype(str).replace('NaN', np.nan)
df = df.dropna()
display(df)
```
